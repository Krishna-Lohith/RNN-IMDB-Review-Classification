```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
      'the glass of juice',
      'the cup of tea',
    'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good',]
```

```python
sent
```

```python
# Define the vocabulary size
voc_size=10000
```

```python
# one hot representation for each word

one_hot_rep = [one_hot(words,voc_size) for words in sent]
one_hot_rep
    
```

```python
# Word embedding representation
from tensorflow.keras.layers import Embedding # to convert the one hot representation into word embedding representation
from tensorflow.keras.utils import pad_sequences # to make the length of all the sentences same
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
set_len = 8 # to make the length of all the sentences same, we should keep more than the length of the longest sentence
embeded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=set_len) # pre means that the padding will be done at the beginning of the sentence, we can add post if we want to add the padding at the end of the sentence
print(embeded_docs)
```

```python
# feature representation using embedding
dim = 10 # dimension of the embedding vector, we can choose any value for this, but it should be less than the vocabulary size
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=set_len)) # voc_size is the vocabulary size, dim is the dimension of the embedding vector, input_length is the length of the input sequence
model.compile('adam','mse') # adam is the optimizer, mse is the loss function
```

```python
model.summary()
```

```python
model.predict(embeded_docs) # to get the feature representation of the sentences using embedding
```

```python
model.predict(embeded_docs).shape # to get the shape of the feature representation of the sentences using embedding
```

```python
# we got (7, 8, 10) which means that we have 7 sentences, each sentence has 8 words and each word is represented by a 10 dimensional vector using embedding
```
